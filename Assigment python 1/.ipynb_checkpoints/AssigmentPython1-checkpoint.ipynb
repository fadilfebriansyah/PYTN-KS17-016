### Penjelasan Data Set  

Kejahatan di wilayah metropolitan besar, seperti London, terjadi dalam pola yang berbeda. Data ini mencakup jumlah laporan kriminal berdasarkan bulan, wilayah LSOA, dan kategori besar/kecil dari Jan 2008-Des 2016.

Column yang ada pada data set<br>

lsoa_code: kode untuk Area Output Super Bawah di London Raya.<br>
borough: Nama umum untuk wilayah London.<br>
major_category: Kategorisasi kejahatan tingkat tinggi<br>
minor_category: Kategorisasi kejahatan tingkat rendah dalam kategori besar.<br>
value: jumlah kejahatan kategorikal yang dilaporkan setiap bulan di wilayah tertentu<br>
year: Tahun penghitungan yang dilaporkan, 2008-2016<br>
month: Bulan penghitungan yang dilaporkan, 1-12<br>

```python
# Mengambil data hanya untuk tahun 2008 hingga 2016
filtered_data = df[(df['year'] >= 2008) & (df['year'] <= 2016)]

# Menghitung total kejahatan per wilayah
borough_crime_counts = filtered_data.groupby('borough')['value'].sum().reset_index()

# Mengurutkan wilayah-wilayah berdasarkan jumlah kejahatan secara menurun
borough_crime_counts = borough_crime_counts.sort_values(by='value', ascending=False)

# Mengambil lima wilayah teratas yang paling banyak menyumbang kejahatan
top_boroughs = borough_crime_counts.head()

# Membuat bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_boroughs['borough'], top_boroughs['value'], color='skyblue')
plt.title('Top 5 Wilayah dengan Jumlah Kejahatan Tertinggi (2008-2016)')
plt.xlabel('Wilayah')
plt.ylabel('Jumlah Kejahatan')
plt.xticks(rotation=45)
plt.tight_layout()

# Menampilkan bar chart
plt.show()
```

```python
# Menghitung total kejahatan per tahun
crime_per_year = df.groupby('year')['value'].sum().reset_index()

# Mengurutkan data berdasarkan tahun
crime_per_year = crime_per_year.sort_values(by='year')

# Extract tahun dan jumlah kejahatan sebagai array NumPy
Year = crime_per_year['year'].to_numpy()
sums = crime_per_year['value'].to_numpy()

# Membuat line graph
plt.figure(figsize=(12, 6))
plt.plot(Year, sums, color='darkgreen', marker='o', linestyle='-', linewidth=2, markersize=8)
plt.title("Total Kejahatan per Tahun (2008-2019)", fontdict={'family': 'Impact', 'color': 'darkgreen', 'weight': 'bold', 'size': 20})
plt.xlabel('Tahun', color='darkgreen', fontsize=14)
plt.ylabel('Jumlah Kejahatan', color='darkgreen', fontsize=14)
plt.xticks(rotation=-45, fontsize=12, color='green')
plt.yticks(fontsize=12, color='blue')
plt.grid(True, linestyle='--', alpha=0.2)
plt.tight_layout()
plt.show()
```
